# Basic Data Operation

You sold 20 pieces of electronics at the price of 12000. The customer gets a total discount of 10%.

Below are those variables:

1.   price=12000
2.   quantity_sold=20
3.   discount=10%


Using Python code, write the code the find the total money that the customer to pay. (Points : 15)

```python
#Input the variables based on the information above
price=12000
quantity_sold=20
discount=0.1

#Code for calculate the total money
total_money=(price*quantity_sold)-(discount*(price*quantity_sold))

#Code to show result of total money
print(total_money)
```

```python
# Replace customer gender list value (Points : 20)
# gender 1 => Male
# gender 0 => Female 

# Create a looping function to read the value of each index in the list
for i in range(len(customer['gender'])):  # a replace function, if 1 will be replaced to Male and other than 1 will be replaced to Female
    if customer['gender'][i] == 1:
        customer['gender'][i] = "Male"
    elif customer['gender'][i] == 0:
        customer['gender'][i] = "Female"
print(customer['gender'])
```
