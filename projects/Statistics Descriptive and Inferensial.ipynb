## Task 1: Compute mean values and variance for all values

**Informasi Data**

```python
df.info()
```

Berdasarkan output di atas setiap kolom tidak memiliki NaN value dan bertipe integer/float. Karena data semua bertipe integer/float maka perhitungan mean dan varian bisa dilakukan menjadi satu dengan code 


*   `mean = df.mean()` dimana mean adalah nama yang digunakan sebagai pendefinisian dari rata-rata dataframe df
*   `var = df.var()` dimana var adalah nama yang digunakan sebagai pendefinisian dari varian dari dataframe df
*   `print(f"Mean =\n{mean}\nVariance =\n{var}")` dimana `print` digunakan untuk menampilkan output, `f` merupakan fungsi yang digunakan untuk melakukan format tampilan dimana nilai` 'Mean'` berisikan nilai `'mean'` yang telah didefinisikan sebelumnya dan nilai `'Varian'` berisikan nilai`'var'` yang telah didefinisikan sebelumnya. Tanda `\n` digunakan agar kata selanjutnya ditampilkan dibaris selanjutnya.

```python
mean = df.mean()
var = df.var()
print(f"Mean =\n{mean}\nVariance =\n{var}")
```

Nilai varian menandakan persebaran data suatu variabel atau seberapa jauh data tersebar dari nilai rerataan. Nilai varian yang besar menandakan bahwa sebaran data jauh dari nilai rerataan data begitu sebaliknya jika varian kecil menandakan bahwa sebaran data dekat dengan rerataan. 

Berdasarkan output di atas variabel yang memiliki varian tinggi yaitu `'AGE'`, `'BP'`,`'S1'`, `'S2'`, `'S3'`,`'S6'`, dan `'Y'`. Sedangkan variabel lainnya memiliki nilai varian yang tergolong rendah.

## Task 2: Plot boxplots for BMI, BP and Y depending on gender

### 1. Variabel BMI



*   Statistik Deskriptif



Menampilkan statistik deskriptif dari masing-masing variabel BMI berdasarkan gender/Sex dengan code `BMI_by_Sex = df.groupby('SEX')['BMI'].describe()` dimana BMI_by_Sex adalah nama baru untuk data statistik deskriptif, kemudian `df.groupby('SEX')['BMI']` dimaksudkan bahwa pada dataframe df akan mengelompokan variabel `BMI` berdasarkan `SEX`

```python
BMI_by_Sex = df.groupby('SEX')['BMI'].describe()
BMI_by_Sex 
```



*   Boxplot Indeks Massa Tubuh Pasien Diabetes Berdasarkan Sex





Akan digunakan library matplotlib dalam membuat boxplot dengan code

## Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

### 1. Variabel Age

```python
#digunakan library seaborn untuk melihat grafik distribusi pada kolom umur pada dataframe 'df'
sns.displot(data=df['AGE'], kde=True)
```

Berdasarkan grafik distribusi di atas terlihat bahwa umur 50-55 tahun memiliki jumlah pasien diabetes yang paling banyak dan dari grafik terlihat bahwa terdapat skewness ke arah kiri/left (negative skewness) sehingga distribusi dari umur pasien diabetes adalah `left-skewed distribution`

### 2. Variabel Sex

```python
#digunakan library seaborn untuk melihat grafik distribusi pada kolom jenis kelamin pada dataframe 'df'
sns.displot(data=df['SEX'], kde=True)
```

Misalkan 1=Laki-laki dan 2=Perempuan

Berdasarkan grafik distribusi di atas terlihat jumlah pasien diabetes laki-laki mendominasi dan dari grafik terlihat bahwa SEX merupakan variabel data bernilai kategorik dengan 2 kemungkinan nilai data maka data berdistribusi secara `Binomial`.

### 3. Variabel BMI

```python
#digunakan library seaborn untuk melihat grafik distribusi pada kolom BMI pada dataframe 'df'
sns.displot(data=df['BMI'], kde=True)
```

Berdasarkan grafik distribusi di atas terlihat bahwa banyak pasien diabetes yang memiliki massa tubuh berkisar 24-27 dan dari grafik terlihat bahwa terdapat skewness ke arah kanan/right (positive skewness) sehingga distribusi dari umur pasien diabetes adalah `right-skewed distribution`

### 4. Variabel Y

```python
#digunakan library seaborn untuk melihat grafik distribusi pada kolom Y pada dataframe 'df'
sns.displot(data=df['Y'], kde=True)
```

Berdasarkan grafik distribusi di atas terlihat bahwa banyak pasien diabetes yang memiliki perkembangan selama 1 tahun berkisar 52-58 dan dari grafik terlihat bahwa terdapat skewness ke arah kanan/right (positive skewness) sehingga distribusi dari umur pasien diabetes adalah `right-skewed distribution`
