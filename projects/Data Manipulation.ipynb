# Import Library

```python
#import libraries here
import warnings #untuk mengignore informasi
warnings.filterwarnings('ignore')

import numpy as np #untuk array
import pandas as pd #untuk dataframe
import seaborn as sns #untuk visulisasi data
import matplotlib.pyplot as plt #untuk visualisasi data
%matplotlib inline 
```

Berikut adalah data frame pertama yang disebut `df1`dengan bantuan pandas `pd.DataFrame` kemudian masukan dictionary yang ingin dibuat data frame.

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df1 #untuk menampilkan dataframe df1
                   
```

Berikut adalah data frame kedua yang disebut `df2`dengan bantuan pandas `pd.DataFrame` kemudian masukan dictionary yang ingin dibuat data frame.

```python
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
df2 #untuk menampilkan dataframe df2
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

Berikut adalah data frame yang disebut `df3` dimana akan menggabungakan data yang sama-sama ada di `df1` dan `df2` berdasarkan `'key'` dengan code `df3=pd.merge(df1,df2,how='inner',on='key')` dimana `pd.merge` adalah fungsi pandas yang digunakan kemudian isi dengan data frame yang akan digabungkan yaitu df1,df2, `how='inner'`menandakan bahwa digabungkan berdasarkan irisan kedua data frame dan code `on='key'` untuk menentukan dataframe akan terbentuk berdasarkan kolom 'key'

```python
#code goes here
df3=pd.merge(df1,df2,how='inner',on='key')
df3 #untuk menampilkan dataframe df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

Akan dibuat data frame yang disebut `df4` dimana akan menggabungakan data yang ada di `df1` dan `df2` berdasarkan `'key'` dengan code `df4=pd.merge(df1,df2,how='left',on='key')` dimana `pd.merge` adalah fungsi pandas yang digunakan kemudian isi dengan data frame yang akan digabungkan yaitu df1,df2, `how='left'`menandakan bahwa dataframe df1 akan diletakan diawal dan ditampilkan semua sedangkan df2 akan menyesuaikan sesuai key, misalnya pada 'key' K3 tidak ada nilai pada B maka akan terisi NaN dan seterusnya, kemudian code `on='key'` untuk menentukan dataframe akan terbentuk berdasarkan kolom 'key'

```python
#code goes here
df4=pd.merge(df1,df2,how='left',on='key')
df4 #untuk menampilkan dataframe df4
```

1.   Akan diganti value K2 menjadi K3 pada df1 dengan code `df1.replace(to_replace={'key':'K2'}, value={'key':'K3'}, inplace = True)` dimana df1 adalah dataframe yang ingin diubah, replace adalah fungsi pandas yang digunakan untuk melakukan pergantian value, `to_replace` merupakan fungsi untuk memanggil value yang akan dirubah yaitu value K2 dalam key dan `value=` digunakan untuk menentukan value akan diganti menjadi apa yaitu K3 dalam keys 'key', dan `inplace = True` berfungsi untuk memproses langsung dataframe yang ingin dirubah tanpa membuat nama baru

```python
df5=pd.merge(df1,df2,how='left',on='key')
df5 #untuk menampilkan dataframe df5
```

Create a dictionary and convert it to a dataframe called "Customer_df" (15 pts)


name=['Anna','Jason','Cindy']

Akan dibuat dictionary yang diberi nama dict_1 yang berisi 3 keys yaitu name, age, dan gender. Value pada key `'name'` yaitu ['Anna','Jason','Cindy'], keys `'age'` yaitu [22,23,21], dan keys `'gender'` yaitu ['F','M','F']. Untuk nama keys maupun value merupakan string maka diapit oleh '...' atau "..." kecuali untuk values keys `'age'` tidak perlu diapit '..' karena bentuknya integer dan setiap key maupun value dipisahkan dengan tanda koma (,).

Setelah dibuat dictionary (dict_1), akan dibuat kedalam dataframe yang disebut Customer_df dengan code `Customer_df=pd.DataFrame(dict_1)` dimana `pd.DataFrame`adalah fungsi pandas untuk membuat dataframe dan diikuti (dictionary yang akan diubah ke dataframe).

```python
#code goes here
dict_1={
    'name':['Anna','Jason','Cindy'],
    'age':[22,23,21],
    'gender':['F','M','F']
}
Customer_df=pd.DataFrame(dict_1)
Customer_df #untuk menampilkan dataframe Customer_df
```

### Delete the "age" column (5 pts)

Akan dihapus kolom 'age' pada dataframe 'Customer_df' yang akan diberi nama Customer_df1 dengan code `Customer_df1=Customer_df.drop('age',axis=1)` dimana `Customer_df.drop` merupakan fungsi untuk menghapus suatu dataframe kemudian diikuti dengan `'age'` sebagai kolom yang ingin dihapus dan `axis =1` menandakan bahwa kolom yang akan didelete karena jika tidak menggunakan axis=1 code akan error karena tidak tahu bagian apa yang didelete.

```python
#code goes here
Customer_df1=Customer_df.drop('age',axis=1)
Customer_df1 #untuk menampilkan dataframe Customer_df1
```

### Rename the "name" column into "first name" (10 pts)

Akan diganti nama pada kolom 'name' menjadi 'first name' dengan code `Customer_df.rename(columns={'name':'first name'},inplace=True)`dimana `Customer_df` adalah nama dataframe, rename adalah fungsi untuk mengganti nama suatu kolom kemudian diikuti dengan `(column={'nama kolom lama' : 'nama kolom baru'` dan `inplace=True` digunakan agar dapat memproses langsung dataframe yang akan dirubah tanpa membuat nama baru.

```python
#code goes here
Customer_df.rename(columns={'name':'first name'},inplace=True)
Customer_df.head() #untuk menampilkan dataframe Customer_df yang telah diubah nama kolomnya
```

### Sort the dataframe by age, descending (10 pts)

Akan ditampilkan dataframe berdasarkan umur yang paling tua ke yang paling muda dengan code `Customer_df=Customer_df.sort_values(['age'],ascending=False)` dimana `Customer_df` adalah nama dataframe, `Customer_df.sort_values` adalah fungsi yang digunakan untuk memfilter value dari suatu kolom dan diikuti `(['kolom acuan', ascending=False` menandakan data akan diurutkan dari yang paling tinggi)

```python
#code goes here
Customer_df=Customer_df.sort_values(['age'],ascending=False)
Customer_df #untuk menampilkan dataframe Customer_df yang telah diurutkan berdasarkan umur
```

## Using data aggregation in pandas , Answer the question below:

Show the total amount of passengers who survived and not survived (10pts)

Akan ditampilkan jumlah passengers yang 'survived' dan 'not survived' dengan code `df.groupby('Survived').agg({'PassengerId':['count']})` dimana df.groupby('Survived') merupakan fungsi untuk melakukan grouping data pada dataframe df berdasarkan kolom Survived. Pada dataframe 'df' variabel kolom 'PassengerId' memiliki type data kategori. oleh karena itu untuk menjumlahkan banyaknya passengers digunakan operasi `'count'`

```python
#code goes here
df.groupby('Survived').agg({'PassengerId':['count']})
```
